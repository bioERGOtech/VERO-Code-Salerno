```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re
import numpy as np

# -----------------------------
# Import and output paths (yours)
# -----------------------------
df = pd.read_excel(r"/Users/guidoputignano/PycharmProjects/VERO_code_edit/Phase_1/data/processed/cleaned_data.xlsx")
```

```python
df['ast_got_range'] = df['ast_got_range'].replace({
    '<21': 'Normal (<21 U/L)',
    '>=21': 'Elevated (>=21 U/L)'
})
```

```python
# Check if old values still exist
print("Checking for old values in ast_got_range:")
print(df['ast_got_range'].value_counts())

# Or more specifically, check for the exact old values
old_values = ['<21', '>=21']
still_present = df['ast_got_range'].isin(old_values).sum()

if still_present > 0:
    print(f"\nWarning: {still_present} rows still contain old values")
    print(df[df['ast_got_range'].isin(old_values)]['ast_got_range'].value_counts())
else:
    print("\nSuccess: No old values found. All values have been replaced.")
```

```python
df['ast_got_range'] = df['ast_got_range'].replace({
    '<21': 'Normal (<21 U/L)',
    '>=21': 'High (>=21 U/L)',
    '<21; >=21': 'Normal (<21 U/L)'  # Add this line
})

# Verify again
print("Updated value counts:")
print(df['ast_got_range'].value_counts())
```

```python
# Adjust blood_glucose_range categories
df['blood_glucose_range'] = df['blood_glucose_range'].replace({
    'Normal (65-110 mg/dL)': 'Normal (65-110 mg/dL)',
    'Missing': 'Missing',
    'High (>110 mg/dL)': 'High (>110 mg/dL)',  # Add this line if needed
    'Low (<65 mg/dL)': 'Low (<65 mg/dL)',    # Add this line if needed
    '65-110; >110': 'Normal (65-110 mg/dL)'        # Add this line if needed
})

# Verify the changes
print("Updated blood_glucose_range counts:")
print(df['blood_glucose_range'].value_counts())
```

```python
# Fix alt_gpt_range as well
df['alt_gpt_range'] = df['alt_gpt_range'].replace({
    '<21': 'Normal (<21 U/L)',
    '>=21': 'High (>=21 U/L)',
    '<21; >=21': 'Normal (<21 U/L)'
})

# Verify the replacement
print("Checking alt_gpt_range:")
print(df['alt_gpt_range'].value_counts())
```

```python
# Fix alcohol_consumption column
df['alcohol_consumption'] = df['alcohol_consumption'].replace({
    'Non-drinker': 'Non-drinker',
    'Not Known / Missing': 'Missing',
    'Yes Moderate': 'Moderate',
    'Missing': 'Missing',
    'Yes High': 'High'
})

# Verify the replacement
print("Checking alcohol_consumption:")
print(df['alcohol_consumption'].value_counts())
```

```python

OUT_DIR = r"/Users/guidoputignano/PycharmProjects/VERO_code_edit/Phase_1/outputs/plot"

# Make subfolders
os.makedirs(f"{OUT_DIR}/numeric_histograms", exist_ok=True)
os.makedirs(f"{OUT_DIR}/numeric_boxplots", exist_ok=True)
os.makedirs(f"{OUT_DIR}/categorical_barcharts", exist_ok=True)

# -----------------------------
# Config
# -----------------------------
MULTI_ENTRY_EXACT = {
    "active_principle",
    "comobidities",
    "adr_chemo_action",
    "adr_macro_category",
    "adr_clean.1",
}

def safe_name(s: str) -> str:
    return re.sub(r"[^A-Za-z0-9_\-]+", "_", str(s)).strip("_")

def is_date_col_name(col: str) -> bool:
    return str(col).lower().endswith("_date")

def is_multi_entry_col(col: str) -> bool:
    return str(col).strip().lower() in {c.lower() for c in MULTI_ENTRY_EXACT}

# -----------------------------
# Identify columns to include
# -----------------------------
date_cols = [c for c in df.columns if is_date_col_name(c)]
multi_entry_cols = [c for c in df.columns if is_multi_entry_col(c)]

numeric_cols = [
    c for c in df.select_dtypes(include=[np.number]).columns
    if c not in date_cols and c not in multi_entry_cols
]
categorical_cols = [
    c for c in df.select_dtypes(include=["object", "category"]).columns
    if c not in date_cols and c not in multi_entry_cols
]

print("Ignoring date columns (_date):", date_cols)
print("Ignoring multi-entry columns:", multi_entry_cols)
print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)

# -----------------------------
# Plot numeric variables
# -----------------------------
for col in numeric_cols:
    s = df[col].dropna()

    # Histogram
    plt.figure(figsize=(7, 4))
    plt.hist(s, bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, "numeric_histograms", f"{safe_name(col)}_hist.png"), dpi=300)
    plt.close()

    # Boxplot
    plt.figure(figsize=(7, 2.8))
    plt.boxplot(s, vert=False)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, "numeric_boxplots", f"{safe_name(col)}_box.png"), dpi=300)
    plt.close()

# -----------------------------
# Plot categorical variables
# -----------------------------
for col in categorical_cols:
    counts = df[col].astype("object").fillna("NaN").value_counts()
    counts_to_plot = counts.head(30)  # cap high-cardinality for readability

    plt.figure(figsize=(9, 4.5))
    counts_to_plot.plot(kind="bar")
    plt.title(f"Bar chart of {col} (top {len(counts_to_plot)})")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig(os.path.join(OUT_DIR, "categorical_barcharts", f"{safe_name(col)}_bar.png"), dpi=300)
    plt.close()

print("\n✅ All plots saved to:")
print(f"  {OUT_DIR}\\numeric_histograms")
print(f"  {OUT_DIR}\\numeric_boxplots")
print(f"  {OUT_DIR}\\categorical_barcharts")
```
