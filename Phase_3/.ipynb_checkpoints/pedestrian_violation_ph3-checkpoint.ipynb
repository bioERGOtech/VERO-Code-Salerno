In this section, we load data from multiple Excel files and sheets, combine it into a single DataFrame, and perform initial preprocessing. This involves adding metadata columns (`condition` and `location`) to track the source of the data and renaming columns for consistency.

### Steps:
1. **File Paths and Sheet Names**:

2. **Reading and Appending Data**:

## ðŸ“Š Exploratory Data Analysis (EDA)

Here, we perform basic exploratory analysis on the combined dataset to understand its structure and relationships.

### Steps:
1. **Summary Statistics**:
   - Compute and print summary statistics to understand numerical distributions.

2. **Missing Values**:
   - Check for and print the count of missing values in each column.

3. **Visualizations**:
   - Use visual tools to explore relationships and distributions:
     - A **heatmap** to identify correlations among numerical variables.
     - A **bar plot** to visualize violations by gender across conditions.
     - A **box plot** to study the spread of violations across different conditions.

---

### Outputs:
1. **Heatmap of Correlations**:
   - Helps identify relationships between numerical variables like violations and other metrics.
2. **Bar Plot (Gender vs Violations)**:
   - Highlights gender differences in violations across conditions.
3. **Box Plot (Violations by Condition)**:
   - Shows the distribution of violations for each condition.

---



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define file paths for the five locations
file_paths = [
    'Location_1P.xlsx',
    'Location_2P.xlsx',
    'Location_3P.xlsx',
    'Location_4P.xlsx',
    'Location_5P.xlsx'
]

# Define sheet names
sheet_names = ['day_peak', 'day_off', 'weekend_peak', 'weekend_off']

# Read all sheets from all files
data_list = []

for file in file_paths:
    for sheet in sheet_names:
        temp_df = pd.read_excel(file, sheet_name=sheet)
        temp_df['condition'] = sheet  # Add a column for the condition
        temp_df['location'] = file  # Add a column for the location
        data_list.append(temp_df)

# Combine all data into one DataFrame
data = pd.concat(data_list, ignore_index=True)

# Rename columns for consistency
data.rename(columns={'Gender': 'gender', 'Type of violations ': 'violations'}, inplace=True)

# Check dataset structure
print(data.info())

# Exploratory Data Analysis (EDA)
# 1. Summary statistics
print(data.describe())

# 2. Check for missing values
print(data.isnull().sum())

# 3. Visualizations
# Heatmap of correlations - only numeric columns
numeric_data = data.select_dtypes(include=[np.number])  # Select only numeric columns

if not numeric_data.empty:
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
    plt.title("Heatmap of Correlations Across Locations and Conditions")
    plt.show()

# Barplot for Gender vs Violations Across Conditions
plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='violations', hue='condition', data=data)
plt.title("Gender vs Violations Across Conditions")
plt.show()

# Boxplot for Violations by Condition Across Locations
plt.figure(figsize=(10, 6))
sns.boxplot(x='condition', y='violations', data=data)
plt.title("Violations by Condition Across Locations")
plt.show()
```



## ðŸ“ˆ Statistical Modeling: Multivariate Regression

In this section, we perform a **multivariate regression** to analyze the relationship between multiple independent variables and the dependent variable, `violations`.

### Steps:
1. **Feature Selection**:
   - Select the independent variables (`Age` and `gender`) as `X_reg`.
   - Use the dependent variable (`violations`) as `y_reg`.

2. **Model Initialization**:
   - Initialize a linear regression model using `LinearRegression()`.

3. **Model Training**:
   - Fit the regression model to the data using the `fit()` method.

4. **Output Regression Parameters**:
   - Print the regression coefficients for each feature, which indicate their influence on the dependent variable.
   - Print the intercept, representing the baseline value of `violations` when all predictors are zero.

---

### Outputs:
- **Regression Coefficients**:
  - Indicate how much the dependent variable changes for a one-unit change in each independent variable.
- **Intercept**:
  - The starting point of the regression line.



```python
# Statistical Modeling
# Multivariate Regression
X_reg = data[['Age ', 'gender']]
y_reg = data['violations']
reg_model = LinearRegression()
reg_model.fit(X_reg, y_reg)
print(f"Regression Coefficients: {reg_model.coef_}")
print(f"Intercept: {reg_model.intercept_}")
```



## ðŸ“ˆ Statistical Modeling: Logistic Regression

In this section, we perform **logistic regression** to predict binary outcomes (violations) based on selected predictors.

### Steps:
1. **Feature and Target Selection**:
   - Independent variables (`X_log`): Select `Age`, `gender`, and `Time` as features.
   - Dependent variable (`y_log`): Use `violations` to predict a binary outcome.

   > Note: The binary target variable (`violations`) can be adjusted for binary classification (e.g., `violations > 0`).

2. **Data Splitting**:
   - Split the dataset into training and testing sets using `train_test_split()`.
   - Use a test size of 30% and set `random_state=42` for reproducibility.

3. **Model Initialization and Training**:
   - Initialize a logistic regression model using `LogisticRegression()`.
   - Train the model on the training set (`X_train` and `y_train`).

4. **Prediction**:
   - Predict outcomes on the test set (`X_test`) using the trained model.

5. **Evaluation**:
   - Generate a classification report to evaluate the model's performance, including metrics like precision, recall, and F1-score.

---

### Outputs:
1. **Classification Report**:
   - Provides a detailed breakdown of the modelâ€™s performance for each class (e.g., 0 and 1).
   - Key metrics:
     - **Precision**: Correct predictions as a percentage of all predictions for a class.
     - **Recall**: Correct predictions as a percentage of all actual instances of a class.
     - **F1-Score**: Harmonic mean of precision and recall.



```python
# Logistic Regression
X_log = data[['Age ', 'gender', 'Time ']]
y_log = data['violations']
X_train, X_test, y_train, y_test = train_test_split(X_log, y_log, test_size=0.3, random_state=42)
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
y_pred = log_model.predict(X_test)
print("Logistic Regression")
print(classification_report(y_test, y_pred))
```



## ðŸŒ² Statistical Modeling: Random Forest Classification

In this section, we use a **Random Forest Classifier** to predict the binary target variable (`violations`). Random Forest is an ensemble learning method that combines multiple decision trees to improve prediction accuracy and control overfitting.

---

### Steps:

1. **Model Initialization**:
   - Initialize a `RandomForestClassifier` with the following parameters:
     - `n_estimators=100`: Specifies the number of trees in the forest.
     - `random_state=42`: Ensures reproducibility of results.

2. **Model Training**:
   - Fit the Random Forest model on the training data (`X_train`, `y_train`).

3. **Prediction**:
   - Predict the target variable for the test dataset (`X_test`) using the trained model.

4. **Evaluation**:
   - Evaluate the model using a **classification report**, which provides metrics such as precision, recall, and F1-score for each class.

---

### Outputs:
1. **Classification Report**:
   - A breakdown of the Random Forest model's performance, including:
     - **Precision**: The percentage of correct positive predictions.
     - **Recall**: The ability to identify all positive instances.
     - **F1-Score**: The balance between precision and recall.
     - **Support**: The number of occurrences for each class.



```python
# Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
print("Random Forest")
print(classification_report(y_test, y_pred))
```



## ðŸ’» Statistical Modeling: Support Vector Machine (SVM) Classification

In this section, we utilize a **Support Vector Machine (SVM)** to classify the target variable (`violations`). SVMs are powerful for both linear and non-linear classification tasks, with kernel functions enabling the latter.

---

### Steps:

1. **Model Initialization**:
   - Initialize the SVM model with the following parameters:
     - `kernel='rbf'`: Specifies the radial basis function (RBF) kernel for nonlinear classification.
     - `random_state=42`: Ensures reproducibility of results.

2. **Model Training**:
   - Train the SVM model using the training data (`X_train`, `y_train`).

3. **Prediction**:
   - Predict the target variable for the test data (`X_test`) using the trained SVM model.

4. **Evaluation**:
   - Evaluate the model's performance using a **classification report**, which provides metrics such as precision, recall, and F1-score.

---

### Outputs:
1. **Classification Report**:
   - Key metrics for assessing the SVM model's classification performance:
     - **Precision**: Measures how many positive predictions were correct.
     - **Recall**: Measures how well the model identifies all positive instances.
     - **F1-Score**: The harmonic mean of precision and recall.
     - **Support**: The number of actual occurrences for each class in the dataset.



```python
# Support Vector Machine Model
svm_model = SVC(kernel='rbf', random_state=42)  # Use 'rbf' kernel for nonlinear classification
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print("Support Vector Machines")
print(classification_report(y_test, y_pred))
```

In this section, we perform **K-means clustering** to group pedestrians into different clusters based on `Age` and `violations`. Clustering helps identify patterns and relationships within the data without needing labeled outcomes.

---

### Steps:

1. **Setting Environment Variables**:
   - Set `OMP_NUM_THREADS=1` to limit the number of OpenMP threads, which can help improve performance in certain parallel computing environments.

2. **Feature Scaling**:
   - Standardize the features (`Age` and `violations`) using `StandardScaler`. This scales the data to have a mean of 0 and a standard deviation of 1, ensuring that each feature contributes equally to the clustering process.

3. **K-means Clustering**:
   - Initialize a `KMeans` model with `n_clusters=3` to form 3 distinct clusters.
   - Use `fit_predict()` to perform the clustering and assign the resulting cluster labels to the `data` DataFrame.

4. **Visualization**:
   - Use a **scatter plot** to visualize the clusters, with `Age` on the x-axis and `violations` on the y-axis.
   - Color the points by cluster using the `hue` parameter, which helps distinguish the different clusters.

---

### Outputs:
1. **Scatter Plot**:
   - The plot shows the clustering results, with points colored according to their assigned cluster.
   - It provides insights into how `Age` and `violations` are distributed across different clusters, revealing patterns in pedestrian behavior.



```python
import os
os.environ['OMP_NUM_THREADS'] = '1'

# Cluster Analysis using K-means
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['Age ', 'violations']])
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(scaled_data)
sns.scatterplot(x='Age ', y='violations', hue='cluster', data=data, palette='viridis')
plt.title("Clustering of Pedestrian Behavior Across Locations and Conditions")
plt.show()
```
