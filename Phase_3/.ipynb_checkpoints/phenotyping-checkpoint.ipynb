```python
import numpy as np
import pandas as pd
from pathlib import Path
import re

# Paths (unchanged from your structure)
INTEGRATED_PATH = r"C:/Users/HP/OneDrive/Desktop/VERO_code/Phase_3/phase3_outputs/phase3_integrated_data.csv"
TYPES_PATH = r"C:/Users/HP/OneDrive/Desktop/VERO_code/Phase_2/results/tables/features_data_types.xlsx"
OUT_DIR = Path(r"C:/Users/HP/OneDrive/Desktop/VERO_code/Phase_3/phase3_outputs")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Load data
df = pd.read_csv(INTEGRATED_PATH)
types_raw = pd.read_excel(TYPES_PATH)

# Identify patient ID column
id_candidates = [c for c in df.columns if re.search(r"(patient.*id|id.*patient|^id$)", c, flags=re.I)]
id_col = id_candidates[0] if id_candidates else None

# Remove outcome variables from clustering features
exclude = [id_col, "survival_days", "Overall_Survival"]
feature_cols = [c for c in df.columns if c not in exclude]

# Normalize names for type matching
types = types_raw.rename(columns={"Feature": "feature_orig"}) if "Feature" in types_raw.columns else types_raw.copy()
types["feature_norm"] = types.get("feature_orig", types.iloc[:,0]).astype(str).str.strip().str.lower()

df_cols_norm = pd.Series(df.columns, index=df.columns).astype(str).str.strip().str.lower()

# Build dtype map
dtype_map = {}
for _, row in types.iterrows():
    norm = row["feature_norm"]
    match = [c for c in df.columns if c.strip().lower() == norm]
    if match:
        col = match[0]
        dt = str(row.get("Data Type", ""))
        dtype_map[col] = "numeric" if re.search(r"(float|int|double)", dt, flags=re.I) else "categorical"

for c in feature_cols:
    if c not in dtype_map:
        dtype_map[c] = "numeric" if pd.api.types.is_numeric_dtype(df[c]) else "categorical"

num_cols = [c for c in feature_cols if dtype_map[c] == "numeric"]
cat_cols = [c for c in feature_cols if dtype_map[c] == "categorical"]

print(f"Using {len(num_cols)} numeric features and {len(cat_cols)} categorical features.")
```

```python
# Step 7 - Gower Distance (robust, finite, symmetric)
import numpy as np
import pandas as pd
from pathlib import Path

# Assumes df (post-pruning), num_cols, cat_cols, OUT_DIR already exist in your notebook.

def _minmax_scale_safe(a):
    a = a.astype(float)
    col_min = np.nanmin(a, axis=0)
    col_max = np.nanmax(a, axis=0)
    denom = (col_max - col_min)
    denom[denom == 0] = 1.0
    return (a - col_min) / denom

def compute_gower_safe(df_num: pd.DataFrame, df_cat: pd.DataFrame) -> np.ndarray:
    n = len(df_num) if df_num is not None else len(df_cat)
    D = np.zeros((n, n), dtype=float)
    contributed_dims = 0

    # Numeric block
    if df_num is not None and df_num.shape[1] > 0:
        Xn = df_num.copy().astype(float).values
        # median impute
        med = np.nanmedian(Xn, axis=0)
        inds = np.where(np.isnan(Xn))
        if inds[0].size > 0:
            Xn[inds] = np.take(med, inds[1])
        # scale to [0,1]
        Xn = _minmax_scale_safe(Xn)
        # accumulate L1 mean distance
        for i in range(n):
            D[:, i] += np.abs(Xn - Xn[i]).mean(axis=1)
        contributed_dims += df_num.shape[1]

    # Categorical block
    if df_cat is not None and df_cat.shape[1] > 0:
        Xc = df_cat.astype(object).copy()
        Xc = Xc.fillna("__NA__").astype(str).values
        # mismatch per column
        for j in range(Xc.shape[1]):
            col = Xc[:, j]
            mismatch = (col.reshape(-1, 1) != col.reshape(1, -1)).astype(float)
            D += mismatch
        contributed_dims += df_cat.shape[1]

    # Normalize by total contributed dimensions
    if contributed_dims > 0:
        D = D / contributed_dims

    # Force symmetry and proper diagonal
    D = 0.5 * (D + D.T)
    np.fill_diagonal(D, 0.0)

    # Replace any non-finite leftovers
    mask = ~np.isfinite(D)
    if mask.any():
        # Simple, safe replacement: use column medians of finite entries
        col_meds = np.nanmedian(np.where(np.isfinite(D), D, np.nan), axis=0)
        for j in range(D.shape[1]):
            D[mask[:, j], j] = col_meds[j] if np.isfinite(col_meds[j]) else 0.0

    # Clip to [0,1] for safety
    D = np.clip(D, 0.0, 1.0)
    return D

# Build numeric and categorical matrices from your current df
df_work = df.copy()

num_cols_final = [c for c in num_cols if c in df_work.columns]
cat_cols_final = [c for c in cat_cols if c in df_work.columns]

df_num = df_work[num_cols_final] if len(num_cols_final) > 0 else None
df_cat = df_work[cat_cols_final] if len(cat_cols_final) > 0 else None

D_final = compute_gower_safe(df_num, df_cat)

# Hard sanity checks
assert np.isfinite(D_final).all(), "Distance matrix contains non-finite values."
assert np.allclose(np.diag(D_final), 0.0), "Diagonal must be zero."
assert np.max(np.abs(D_final - D_final.T)) < 1e-10, "Distance matrix must be symmetric."

np.save(OUT_DIR / "gower_distance_final.npy", D_final)
print("Gower distance saved:", OUT_DIR / "gower_distance_final.npy", "shape:", D_final.shape)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import pandas as pd
import numpy as np

def cluster_with_k(distance_matrix, k):
    model = AgglomerativeClustering(
        n_clusters=k,
        affinity="precomputed",
        linkage="average"
    )
    labels = model.fit_predict(distance_matrix)
    return labels

# Try K=2 (your best silhouette) and K=3 as a fallback if tiny cluster appears
labels_k2 = cluster_with_k(D_final, 2)
sizes_k2 = pd.Series(labels_k2).value_counts().sort_index()
print("K=2 sizes:\n", sizes_k2)

# Silhouette using precomputed distance (works in recent sklearn; otherwise returns NaN gracefully)
try:
    sil_k2 = silhouette_score(D_final, labels_k2, metric="precomputed")
except Exception:
    sil_k2 = np.nan
print("K=2 silhouette(precomputed):", sil_k2)

# If you see a single item cluster, consider K=3
labels_k3 = cluster_with_k(D_final, 3)
sizes_k3 = pd.Series(labels_k3).value_counts().sort_index()
print("\nK=3 sizes:\n", sizes_k3)
try:
    sil_k3 = silhouette_score(D_final, labels_k3, metric="precomputed")
except Exception:
    sil_k3 = np.nan
print("K=3 silhouette(precomputed):", sil_k3)

# Choose final K
if sizes_k2.min() <= 2 and sizes_k3.min() > 2:
    labels_final = labels_k3
    chosen_k = 3
else:
    labels_final = labels_k2
    chosen_k = 2

print(f"\nChosen K: {chosen_k}")
print(pd.Series(labels_final).value_counts().sort_index())

# Export phenotype labels
labels_out = df[[id_col]].copy() if id_col else pd.DataFrame(index=df.index)
if id_col:
    labels_out[id_col] = df[id_col].values
labels_out["phenotype"] = labels_final
labels_out.to_csv(OUT_DIR / "phenotype_labels.csv", index=False)
print("Saved:", OUT_DIR / "phenotype_labels.csv")
```

```python
# ================================
# STEP 7 (CORRECTED) — Embeddings on clean main cohort
# ================================

import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Use df_main and final K=2 labels
labels_for_plot = df_main["Phenotype_K2"].values

# Rebuild matrices from cleaned data
num_df = df_main[num_cols_kept].copy() if len(num_cols_kept) > 0 else None
cat_df = df_main[cat_cols_kept].copy() if len(cat_cols_kept) > 0 else None

# Impute numeric
if num_df is not None:
    for c in num_cols_kept:
        num_df[c] = pd.to_numeric(num_df[c], errors="coerce")
    num_df = num_df.fillna(num_df.median())

# Impute categorical
if cat_df is not None:
    cat_df = cat_df.astype(str).fillna("__NA__")

# Scale and One-Hot Encode
blocks = []
if num_df is not None and num_df.shape[1] > 0:
    scaler = StandardScaler()
    X_num = scaler.fit_transform(num_df.values.astype(float))
    blocks.append(X_num)

if cat_df is not None and cat_df.shape[1] > 0:
    oh = OneHotEncoder(handle_unknown="ignore", sparse=False)
    X_cat = oh.fit_transform(cat_df.values)
    blocks.append(X_cat)

X_mix = np.hstack(blocks)
print("Embedding matrix:", X_mix.shape)

# Plot helper
def _scatter_2d(Z, labels, title, out_png):
    plt.figure(figsize=(6,5))
    for k in np.unique(labels):
        sel = labels == k
        plt.scatter(Z[sel, 0], Z[sel, 1], s=12, alpha=0.75, label=f"Phenotype {k}")
    plt.title(title)
    plt.legend()
    plt.savefig(Path(OUT_DIR) / out_png, dpi=150)
    plt.show()

# PCA
pca = PCA(n_components=2, random_state=7)
Z_pca = pca.fit_transform(X_mix)
_scatter_2d(Z_pca, labels_for_plot, "PCA 2D (Clean Cohort)", "pca_clean.png")

# t-SNE
tsne = TSNE(n_components=2, random_state=7, init="pca", learning_rate="auto")
Z_tsne = tsne.fit_transform(X_mix)
_scatter_2d(Z_tsne, labels_for_plot, "t-SNE 2D (Clean Cohort)", "tsne_clean.png")
```

```python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Merge df_main with outcome data as described earlier
# Ensure 'patient_id_column' is replaced with the actual column name
OUTCOME_PATH = r"C:/Users/HP/OneDrive/Desktop/VERO_code/Phase_2/results/tables/outcome.csv"
df_outcome = pd.read_csv(OUTCOME_PATH)
merged_df = pd.merge(df_main, df_outcome, on='patient_id', how='left')

# Filter out missing data and select relevant columns
df_km = merged_df[['survival_days', 'Overall_Survival', 'Phenotype_K2']].dropna()

# Kaplan-Meier estimator
kmf = KaplanMeierFitter()

# 1. **Median Survival Comparison**:
plt.figure(figsize=(8, 6))

for phenotype in df_km['Phenotype_K2'].unique():
    # Filter data for each phenotype
    phenotype_data = df_km[df_km['Phenotype_K2'] == phenotype]
    
    # Fit the Kaplan-Meier estimator
    kmf.fit(durations=phenotype_data['survival_days'], event_observed=phenotype_data['Overall_Survival'], label=f"Phenotype {phenotype}")
    
    # Plot the Kaplan-Meier curve for each phenotype
    kmf.plot_survival_function()

    # Print the median survival time for the current phenotype
    median_survival = kmf.median_survival_time_
    print(f"Phenotype {phenotype}: Median Survival Time = {median_survival} days")

plt.title("Kaplan-Meier Survival Curve by Phenotype")
plt.xlabel("Survival Days")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True)
plt.show()

# 2. **Death/Progression/Toxicity Differences**:


# Assuming that 'Overall_Survival' = 1 represents "death"
death_counts = df_km.groupby('Phenotype_K2')['Overall_Survival'].value_counts(normalize=True).unstack(fill_value=0)

# Print the normalized (percentage) distribution of death across phenotypes
print("Death Rate Comparison Across Phenotypes:")
print(death_counts)


# You can perform a statistical test like the log-rank test to formally compare survival distributions:
from lifelines.statistics import logrank_test

# Compare two phenotypes using the log-rank test
phenotype_1 = df_km[df_km['Phenotype_K2'] == 0]  
phenotype_2 = df_km[df_km['Phenotype_K2'] == 1]  

# Perform the log-rank test
results = logrank_test(phenotype_1['survival_days'], phenotype_2['survival_days'],
                       event_observed_A=phenotype_1['Overall_Survival'], event_observed_B=phenotype_2['Overall_Survival'])

# Output the p-value of the comparison
print(f"Log-rank test p-value: {results.p_value}")
```
